## Import all the libraries

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import sqlite3    ## SQL Interface
import pickle     ## Used to save your data - Converts objects to byte stream and vice versa
```

```python
from sklearn.utils import shuffle
```

### Read the updated sqlite

```python
conn = sqlite3.connect('final.sqlite')

final = pd.read_sql_query("""SELECT * FROM Reviews""", conn)

final.head()
```

```python
final["Score"].value_counts()
```

### creating data frame of positive reviews

```python
pos = final[(final['Score'] == 'positive')]
```

```python
pos.shape
```

### creating data frame of negative reviews

```python
neg = final[(final['Score'] == 'negative')]
neg.shape
```

### shuffle the positive data frame

```python
pos = shuffle(pos, random_state = 123, replace = True)
pos.head()
```

```python
pos.shape
```

```python
final.drop(['index','level_0'], axis = 1, inplace = True)
```

```python
pos.drop(['index','level_0'], axis = 1, inplace = True)
neg.drop(['index','level_0'], axis = 1, inplace = True)
```

```python
pos.shape
```

```python
neg.shape
```

```python
df = pos.iloc[:57110,:]
```

```python
df.shape
```

#### concat the data frames

```python
df = pd.concat([df, neg], ignore_index = True)
df
```

```python
df['Score'].value_counts()
```

```python
conn = sqlite3.connect('balanced.sqlite')
c=conn.cursor()
df.to_sql('Reviews', conn, if_exists='replace', index = False)
conn.close()
```
